# 예측 (4가지 방법)

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error,r2_score
```

### 1. LinearRegression

##### 15일

```python
pred_df_B_3_3m_LinearRegression = pred_df_B_3_3m[pred_df_B_3_3m.columns]

model = LinearRegression()


train_data = pred_df_B_3.iloc[1:151]
train_x = train_data[input_var]


N = 15


for i in range(N) :
    
    test_data = pred_df_B_3_3m_LinearRegression.iloc[i:(i+1)]
    test_x = test_data[input_var]
    

    for j in ['B_sales','B_fb_cost', 'B_brand_fb_cost', 'B_brand_views'] :

        train_y = train_data[j]

        
        model.fit(train_x, train_y)
        
        predicted_y = model.predict(test_x)

        pred_df_B_3_3m_LinearRegression[j].iloc[i] = predicted_y
        
        
        if i < N-1 :
            pred_df_B_3_3m_LinearRegression['lag_1_'+j].iloc[(i+1)] = predicted_y

```

```python
pred_y_LinearRegression = pred_df_B_3_3m_LinearRegression.iloc[0:15]['B_sales'].values
real_y = df_B.iloc[151:166]['B_sales'].values

mse = mean_squared_error(real_y, pred_y_LinearRegression)

print()
print("LinearRegression : ")
print('mse : {:.4f}'.format(mse))
print('rmse : {:.4f} '.format(mse**0.5))

print('')
fig, ax = plt.subplots(figsize=(5, 3))
plt.plot(pred_y_LinearRegression, color = 'red', label = 'prediction')
plt.plot(real_y, color = 'green', label = 'real')
plt.ylim([0, 100])     # Y축의 범위: [ymin, ymax]
plt.grid(True)
plt.legend()
plt.show()
```
