## 휴일 여부 추가 (주말 + 공휴일)

```python
import holidays
```

```python
is_holiday = []
```

```python
for date in df_A_index['date'] :
    # 공휴일
    if date in holidays.KOR() :
        is_holiday.append(1)
    
    # 주말
    elif date.weekday() >= 5 :
        is_holiday.append(1)
    
    # 그외    
    else :
        is_holiday.append(0)
        
```

```python
is_holiday
```

```python
len(df_A_index)
```

```python
len(is_holiday)
```

```python
df_A_index['is_holiday'] = is_holiday
```

```python
pd.set_option('display.max.colwidth', 25)
df_A_index
```

## 주말 여부 추가

```python
is_weekend = []
```

```python
for date in df_A_index['date'] :
    if date.weekday() >= 5 :
        is_weekend.append(1)
       
    else :
        is_weekend.append(0)
```

```python
df_A_index['is_weekend'] = is_weekend
```

## 공휴일 여부 추가

```python
is_public_holiday = []
```

```python
for date in df_A_index['date'] :
    if date in holidays.KOR() :
        is_public_holiday.append(1)
        
    else :
        is_public_holiday.append(0)
```

```python
df_A_index['is_public_holiday'] = is_public_holiday
```
