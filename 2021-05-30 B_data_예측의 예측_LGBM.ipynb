### 예측의 예측 진행

```python
pred_df_B_3_3m_LightGBM = pred_df_B_3_3m[pred_df_B_3_3m.columns]

# 예측할 날짜
N = 31


for i in range(N) :
    
    test_data = 0
    
    test_data = pred_df_B_3_3m_LightGBM .iloc[i:(i+1)]
    
    test_x = test_data[input_var]

    
    for j in ['B_sales','B_fb_cost', 'B_brand_fb_cost', 'B_brand_views'] :
        
        test_y = 0

        if j == 'B_sales' :
            test_y = clf_B_sales.predict(test_x, num_iteration = clf_B_sales.best_iteration)

        elif j == 'B_fb_cost' :
            test_y = clf_B_fb_cost.predict(test_x, num_iteration = clf_B_fb_cost.best_iteration)
            
        elif j == 'B_brand_fb_cost' :
            test_y = clf_B_brand_fb_cost.predict(test_x, num_iteration = clf_B_brand_fb_cost.best_iteration)
            
        elif j == 'B_brand_views' :
            test_y = clf_B_brand_views.predict(test_x, num_iteration = clf_B_brand_views.best_iteration)

        
        pred_df_B_3_3m_LightGBM[j].iloc[(i)] = test_y[0]

        if i < N-1 :
            pred_df_B_3_3m_LightGBM['lag_1_'+j].iloc[(i+1)] = test_y[0]
```

### RMSE 구하기

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
pred_y_LightGBM = pred_df_B_3_3m_LightGBM['B_sales'].values
real_y = df_B.iloc[151:182]['B_sales'].values

mse = mean_squared_error(real_y, pred_y_LightGBM)

print()
print("LightGBM : ")
print('mse : {:.4f}'.format(mse))
print('rmse : {:.4f} '.format(mse**0.5))

print('')
fig, ax = plt.subplots(figsize=(5, 3))
plt.plot(pred_y_LightGBM, color = 'red', label = 'prediction')
plt.plot(real_y, color = 'green', label = 'real')
  # Y축의 범위: [ymin, ymax]
plt.grid(True)
plt.legend()
plt.show()
```
